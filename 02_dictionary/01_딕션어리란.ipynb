#1.딕션어리란
*   사전과 비슷한 방식의 데이터 타입으로, 키와 값이 한 쌍을 이루며 중괄호({ })로 표현
*   딕션어리명 = {키:값, 키:값, 키:값,....}
*   키(key) : 인덱스처럼 데이터의 위치를 알려 주고, 각 데이터에 접근할 수 있는 기준이 되는 문자
*   값(value) : 각 키에 해당하는 데이터






```python
eng = {'break':'깨어지다', 'bring':'가져오다', 'brush':'붓', 'brave':'용감한'}
print(eng)
```

##딕션어리의 키와 값
*   딕션어리명['키']로 데이터(값)에 접근할 수 있다.
*   키를 기준으로 데이터를 구성하기 때문에 문자도 들어간다.
*   존재하지 않는 키를 넣으면 KeyError가 발생
*   같은 키가 여러개 존재할 수 없다. 키가 같으면 가장 마자막 값만 저장





```python
eng = {'break':'깨어지다', 'bring':'가져오다', 'brush':'붓', 'brave':'용감한'}
print(eng['break'])
```

```python
eng = {'break':'깨어지다', 'bring':'가져오다', 'brush':'붓', 'brave':'용감한'}
print(eng['brush'])
print(eng['brave'])
```

```python
eng = {'break':'깨어지다', 'bring':'가져오다', 'brush':'붓', 'brave':'용감한'}
print(eng['brief'])
```
