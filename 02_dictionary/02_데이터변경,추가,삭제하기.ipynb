#2.데이터변경, 추가, 삭제하기
##값 변경하기
*   딕션어리명[키] = 값



```python
# 값 변경하기
menu = {'김밥':2500, '참치김밥': 3500, '라면':3000, '치즈라면': 4000}
menu['참치김밥'] = 3000
print(menu)
```

```python
menu['라면'] = 2500
print(menu)
menu['치즈라면'] = 3500
print(menu)
```

```python
menu = {'김밥':2500, '참치김밥': 3500, '라면':3000, '치즈라면': 4000}
menu['김밥'] += 500
print(menu)
```

```python
menu['참치김밥'] -= 800
print(menu)
```

##데이터 추가하기

*   딕션어리[새로운키] = 값



```python
# 키-값 추기하기
menu = {'김밥':2500, '참치김밥': 3500, '라면':3000, '치즈라면': 4000}
menu['돈가스'] = 6000
print(menu)
```

```python
menu['돈가스'] = 6500
print(menu)
```

```python
menu = {'김밥':2500, '참치김밥': 3500, '라면':3000, '치즈라면': 4000}
name = input('추가할 메뉴는? ')
price = int(input('메뉴의 가격은? '))

menu[name] = price
print(menu)
```

##데이터 삭제하기
*   del 딕션어리[키]



```python
# 키-값 삭제하기
menu = {'김밥':2500, '참치김밥':3500, '라면':3000, '치즈라면':4000}

del menu['치즈라면']
print(menu)
```
