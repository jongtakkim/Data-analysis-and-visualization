#3.다양한 명령어
## 딕션어리에서 키만 꺼내기
*   딕션어리.keys()




```python
# 딕셔너리의 키만 꺼내기 : keys()
mart ={'과자':1500, '아이스크림':1700, '휴지':4800, '고기':10500, '우유':2600}
print(mart.keys())
```

##딕션어리에서 값(value)만 꺼내기
*   딕션어리명.values()




```python
# 딕셔너리의 값만 꺼내기: values()
mart ={'과자':1500, '아이스크림':1700, '휴지':4800, '고기':10500, '우유':2600}
print(mart.values())
```

```python
# 추출한 결괏값을 리스트로 바꾸기
mart ={'과자':1500, '아이스크림':1700, '휴지':4800, '고기':10500, '우유':2600}
print(list(mart.keys()))
```

```python
mart ={'과자':1500, '아이스크림':1700, '휴지':4800, '고기':10500, '우유':2600}
print(list(mart.values()))
```

##딕션어리의 키-값 꺼내기
*   딕션어리명.items()



```python
# 딕셔너리의 키-값 꺼내기 : items()
mart ={'과자':1500, '아이스크림':1700, '휴지':4800, '고기':10500, '우유':2600}
print(mart.items())
```
