#6.리스트 명령어 사용하기
##정렬하기
*   리스트 데이터를 오름차순, 내림차순 정렬
*   리시트명.sort()



```python
# 리스트 원소 정렬하기
food =['피자', '치킨', '스테이크', '된장찌개','바나나', '파스타', '감자칩']

food.sort()
print('정렬 후 : ',food)
```

```python
# 리스트 원소 내림차순(역순)으로 정렬하기
food =['피자', '치킨', '스테이크', '된장찌개','바나나', '파스타', '감자칩']

food.sort(reverse=True)
print('정렬 후 : ',food)
```

```python
spending_6mon =[25, 36, 8, 53, 24, 56]

spending_6mon.sort(reverse=True)
print(spending_6mon[0:2])
```

```python
spending_6mon =[25, 36, 8, 53, 24, 56]

spending_6mon.sort(reverse=True)
print(spending_6mon[-2:])
```

##리스트 데이터의 인덱스값 찾기
*   찾으려는 데이터를 index()의 소괄호 안에 해당 데이터를 넣으면 그 데이터가 어느 위치에 있는지, 즉 인덱스 번호를 알려줌



```python
# 원소의 인덱스 찾기
print(spending_6mon.index(24))
```
