#5. 리스트에 필요한 데이터 꺼내기
##리스트 슬라이싱
*   리스트에서 원하는 부분만 골라내는 것
*   리스트명[시작값:끝값+1] - 시작값부터 끝값까지의 인덱스값을 꺼낸다



```python
food = ['피자','치킨','떡볶이','스테이크','파스타']

print(food[0:3])
```

```python
print(food[2:4])
```

```python
print(food[-2:])
```

```python
spend_6mon =[25, 36, 8, 53, 24, 56]

print('1월부터 3월 지출액 :',spend_6mon[:3])
print('3월부터 5월 지출액 :',spend_6mon[2:5])
print('5월부터 6월 지출액 :',spend_6mon[-2:])
```

##문자열 슬라이싱
*   문자열도 리스트처럼 글자에 인덱스처럼 사용할 수  있음




```python
# 문자열 슬라이싱
a = '나는야 우주최강'
print(a[:3])
```

```python
print(a[-4:])
```

```python
print(a[-4:-2])
```

```python
date = '2021-04-15'

print('연도 :',date[:4])
print('월 :',date[5:7])
print('일 :',date[-2:])
```
