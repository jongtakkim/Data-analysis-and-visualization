#4. 리스트에 데이터 추가하고 삭제하기
## 데이터 추가
*   append() : 리스트에 요소를 추가하는 명령어
*   리스트명.append(추가할 데이터)



```python
food = ['피자','치킨','떡볶이','스테이크','파스타']
food.append('감자칩')
print(food)
```

```python
food.append('카레')
print(food)
```

```python
abc = []
abc.append(24)
abc.append(38)
abc.append(19)
abc.append(40)
print(abc)
```

```python
to_do = []
print('해야할 일을 차례대로 입력해주세요.')

for i in range(5) :
    to_do.append(input())

print(to_do)
```

##데이터 삭제하기
*   리스트명.remove(삭제할데이터)
*   del 리스트명[인덱스]



```python
food = ['피자','치킨','떡볶이','스테이크','파스타']
food.remove('떡볶이')
print(food)
```

```python
food = ['피자','치킨','떡볶이','스테이크','파스타']
del food[-1]
print(food)
```
