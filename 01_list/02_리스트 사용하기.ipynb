# 2. 리스트 사용하기
*   인덱스(index) - 리스트에서요소의 위치를 알려주는 숫자
*   인덱스는 0부터 시작



```python
spending = [25, 36, 8, 53, 24, 56]
print(spending[0])
```

```python
print(spending[3])
```

```python
print(spending[6])
```

```python
food = ['피자','치킨','떡볶이','스테이크','파스타']

# 가장 좋아하는 음식은?
print(food[0])

# 세 번째로 좋아하는 음식은?
print(food[2])

# 가장 안 좋아하는 음식은?
print(food[4])
```

## 리스트의 값 변경하기
*   리스트 안에 있는 개별 요소가 필요할 때는 해당 위치의 인덱스 활용
*   변수에 값을 더하고 빼는 방법과 같음



```python
spending = [25, 36, 8, 53, 24, 56]
print('3월 지출 변경 전:', spending)
spending[2] += 10
print('3월 지출 변경 후:', spending)
```

```python
spending[4] -= 2
print('5월 지출 변경 후:', spending)
```
