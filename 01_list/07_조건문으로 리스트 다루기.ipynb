#7.조건문으로 리스트 다루기
##음식 순위 출력
*   음식을 좋아하는 순서대로 저장한 food 리스트 만들기
*   음식 이름을 입력받아 해당 음식의 순위 출력하기



```python
food =['피자', '치킨', '스테이크', '된장찌개']
search = input('찾는 음식은? ')

if search in food :
    print(food.index(search)+1,'위입니다.')
else :
    print(search,'은/는 순위에 없습니다')
```

##영상 추천 프로그램 만들기
*   시청 내역에 다이어트가 있으면 홈트레이닝 영상을 추천
*   시청 내역에 뉴스가 있으면 뉴스속보 영상을 추천



```python
history = []

# 새로운 종류의 영상 시청
for i in range(5) :
    history.append(input('시청한 영상은? '))
print('지금까지 시청하신 주제는',history)

if '다이어트' in history:
    print('홈트레이닝 영상도 좋아하실 것 같아요!')
if '뉴스' in history :
    print('뉴스를 좋아하시면 속보 영상도 추천합니다!')
```
