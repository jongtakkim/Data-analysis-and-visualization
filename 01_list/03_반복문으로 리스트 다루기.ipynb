#3.반복문으로 리스트 다루기
##리스트명 사용하기
*   반복문에 데이터를 직접 넣는 대신 리스트를 넣어도 됨



```python
for i in [25, 36, 8, 53, 24, 56] :
    print(i)
```

```python
spending = [25, 36, 8, 53, 24, 56]

for s in spending :
    print(s)
```

##인덱스 사용하기
*   인덱스는 0부터 시작
*   리스트 요소 개수가 틀리지 않게 주의



```python
spending = [25, 36, 8, 53, 24, 56]

for i in range(6) :
    print(spending[i])
```

```python
for i in range(6) :
    print(i+1,'월 지출은',spending[i])
```

```python
for i in range(7) :
    print(i+1,'월 지출은',spending[i])
```

```python
for i in range(len(spending)) :
    print(i+1,'월 지출은',spending[i], end=', ')
```

##음수 인덱스 사용하기
*   인덱스에 음수도 사용할 수 있음
*   음수 인덱스는 오늘쪽에서 시작해 왼쪽으로 1씩 감소



```python
food = ['피자','치킨','떡볶이','스테이크','파스타']
# 가장 덜 좋아하는 음식?
print(food[-1])
```

```python
# 두번째로 덜 좋아하는 음식?
print(food[-2])

# 세번째로 덜 좋아하는 음식?
print(food[-3])
```
