#8. 중첩제어구조로 리스트 다루기
## 무작위 점수 만들어서 구분하기
*   score 라는 빈 리스트 만들기
*   score.append()에 넣어 무작위 점수를 score 리스트에 추가하기
*   80점 이상인 값은 A 리스트에 저장히기
*   40점 이상 80점 미만인 값은 B 리스트에 저장하기
*   40점 미만인 값은 C 리스트에 저장하기










```python
import random

score = []
A = []
B = []
C = []

for i in range(30) :
    score.append(random.randint(0, 100))

for i in range(len(score)) :
    if score[i] >= 80 :
        A.append(score[i])
    elif score[i] >= 40 :
        B.append(score[i])
    else :
        C.append(score[i])
print(score)
print(A)
print(B)
print(C)
```
