#9.이차원 리스트
##이차원 리스트란?
*   리스트 안에 리스트가 들어 있는 형태
*   abc = [[1,2,3].[4,5,6].[7,8,9]]
*   이차원 리스트의 인덱스는 2개(행과 열)





```python
abc = [[1,2,3], [4,5,6], [7,8,9]]
print(abc[0])
print(abc[1])
print(abc[2])
```

```python
abc = [[1,2,3], [4,5,6], [7,8,9]]
print(abc[0][1])
print(abc[1][2])
```

##반복문으로 이차원 리스트 다루기

```python
for row in abc :
    print(row)
```

```python
for row in abc :
    print(row[0])
```

```python
for row in abc :
    print(row[-1])
```

##이차원 리스트로 간단한 프로그램 만들기

*   올림픽에서 나라별 순위와 메달의 갯수를 리스트로 작성
*   순위,나라명,금메달수,은메달수,동메달수,총메달수



```python
olympic =[[1,'미국',46,37,38,121],
          [2,'영국',27,23,17,67],
          [3,'중국',26,18,26,70],
          [4,'러시아',19,18,19,56],
          [5,'독일',17,10,15,42],
          [6,'일본',12,8,21,41],
          [7,'프랑스',10,18,14,42],
          [8,'대한민국',9,3,9,21]]
```

```python
print(olympic[0][1],'금메달 개수 :',olympic[0][2])
print(olympic[4][1],'금메달 개수 :',olympic[4][2])
```

```python
print(olympic[:4])
print(olympic[:4][1])
```

```python
for row in olympic :
    print(row[:2])
```

```python
print('금메달보다 은메달이 더 많은 나라')
for row in olympic :
    if row[2] < row[3] :
        print(row[1])
```

```python
print('금메달보다 동메달이 더 많은 나라')
for row in olympic :
    if row[2] < row[4] :
        print(row[1])
```

```python
#총 금매달 갯수
total_gold =0
for row in olympic :
    total_gold += row[2]
print(total_gold)
```

```python
# 총 메달 중 금메달이 차지하는 비율이 40%가 넘는 나라
for row in olympic :
    gold = row[2]/row[-1] *100
    if gold > 40 :
        print(row[1], gold,'%')
```
